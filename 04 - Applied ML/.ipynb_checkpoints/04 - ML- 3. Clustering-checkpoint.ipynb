```python
import pandas as pd
import numpy as np
import os
from IPython.display import Image
from IPython.core.display import HTML 
import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D
% matplotlib inline

import sklearn 
from sklearn.cluster import KMeans
from sklearn import metrics
```

## Loading data from the pre-processing part
We will separate the skin rating variable from the data. However, as we are looking into an unsupervised clustering, we will not use the labels df_y in the clustering algorithm, but rather confront it to the labelling obtained, to check how close the clustering relates to skin color (we will call this feature **score**, as opposed to clusters **silhouette** score).

```python
df_p = pd.read_csv("crowdstorm_players_aggregated.csv", encoding = "utf-8")
df_y = np.asarray(df_p['mean_rater'], dtype=int)
#When we want the data freshly loaded, we will refer to df_x_load, whereas we will make extensive use of df_x later on.
df_x_load = df_p.drop('mean_rater', axis=1, errors='ignore')
```

## Discarding the classification-oriented features

#### After a few experiments on clustering,
we found that some of the features that we introduced in the classification part were problematic in this one, since when eliminating features iteratively down to said features (selecting the feature to remove iteratively in a **greedy manner** with respect to score or silhouette) , the result became completely distorted. 
Indeed, it is obvious that when your features are down to wheter a player is "Right Midfielder" or "Right Winger", you will get a silhouette close to 1 (0.935314994377) even with k=2, given players are either one of both or neither. A plot of the situation shows clearly three separate clusters centered on [0,0], [1,0] and [1,0]. Interestingly enough, this also improves the label score with respect to skin color up to 0.74, but this is *probably* overfitting.

```python
df_x_load.drop('England', inplace=True,axis=1,errors='ignore')
df_x_load.drop('France', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Germany', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Spain', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Attacking Midfielder', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Center Back', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Center Forward', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Center Midfielder', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Defensive Midfielder', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Goalkeeper', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Left Winger', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Right Fullback', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Left Fullback', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Left Midfielder', inplace=True,axis=1,errors='ignore')    
df_x_load.drop('Right Midfielder', inplace=True,axis=1,errors='ignore')
df_x_load.drop('Right Winger', inplace=True,axis=1,errors='ignore') 
```

## Clustering
Here we want to use an unsupervised learning technique to cluster the soccer players in 2 disjoint clusters. We will use the K-Means method for that purpose, which is appropriate to our case (good amount of samples, low number of clusters...), as provided in the scikit-learn library. 

#### First, we create three score-related functions 
**score** gives the score (as defined above) of the current kmeans model, **silhouette** returns the clusters silhouette score, and **print_scores** simply prints out both to the output.

```python
def score(df_y, labels):
    kmeans_score = sum(labels==df_y)/len(df_y)
    return kmeans_score
def silhouette(df_x, labels):
    kmeans_silhouette = metrics.silhouette_score(df_x, labels, metric='euclidean')
    return kmeans_silhouette
def print_scores(df_x, df_y, labels):
    print("Score :", score(df_y, labels))
    print("Silhouette :", silhouette(df_x, labels))
```

We start by creating a simple Kmeans model with two clusters, to see the results of a clustering without feature optimization.
A first observation : randomized initial state will yield the same clusters silhouette, but gives high variance on the label score. We find that using random_state = 0 is therefore more convenient in order to eventually find a set of features showing high score.

```python
#Randomized
kmeans_model_start = KMeans(n_clusters=2).fit(df_x_load)
print_scores(df_x_load, df_y, kmeans_model_start.labels_)
```

```python
#Not randomized
kmeans_model_start = KMeans(n_clusters=2, random_state = 0).fit(df_x_load)
print_scores(df_x_load, df_y, kmeans_model_start.labels_)
```

## Feature optimization

Next, we define two functions that will basically perform the same greedy feature deletion, although one will be looking to improve **silhouette score**, whereas the second will look to improve **label score**.

```python
def remove_feat_silhouette(df_x_load):
    df_x=df_x_load.copy()
    loop_size = len(df_x.columns)
    best_cluster_scores = []
    param_cluster = []

    #Loop for features refining
    cluster_scores = []
    for i in range(0, loop_size -2):
        df_x_temp = df_x.copy()
        df_x_temp.drop(df_x_temp.columns[i],inplace=True,axis=1,errors='ignore')


        # KMeans clustering
        #kmeans_model = KMeans(n_clusters=2, random_state=0).fit_predict(df_x_temp, y=df_y)
        kmeans_model = KMeans(n_clusters=2, random_state=0).fit(df_x_temp)
        #Scores
        cluster_scores.append(silhouette(df_x_temp, kmeans_model.labels_))
    
    best_cluster_scores.append(np.max(cluster_scores))
    param_cluster.append(df_x.columns[np.argmax(cluster_scores)])
    #print('best score:', best_cluster_scores[-1],'best feature to delete:', param_cluster[-1] )
    
    df_x.drop(df_x.columns[np.argmax(cluster_scores)], inplace=True,axis=1,errors='ignore')
    return df_x

def remove_feat_score(df_x_load):
    df_x=df_x_load.copy()
    loop_size = len(df_x.columns)
    best_cluster_scores = []
    param_cluster = []

    #Loop for features refining
    cluster_scores = []
    for i in range(0, loop_size -2):
        df_x_temp = df_x.copy()
        df_x_temp.drop(df_x_temp.columns[i],inplace=True,axis=1,errors='ignore')


        # KMeans clustering
        #kmeans_model = KMeans(n_clusters=2, random_state=0).fit_predict(df_x_temp, y=df_y)
        kmeans_model = KMeans(n_clusters=2, random_state=0).fit(df_x_temp)
        #Scores
        cluster_scores.append(score(df_y, kmeans_model.labels_))
    
    best_cluster_scores.append(np.max(cluster_scores))
    param_cluster.append(df_x.columns[np.argmax(cluster_scores)])
    #print('best score:', best_cluster_scores[-1],'best feature to delete:', param_cluster[-1] )

    
    df_x.drop(df_x.columns[np.argmax(cluster_scores)], inplace=True,axis=1,errors='ignore')
    return df_x
```

Now, we will run these methods on our dataset until only one feature is left, to get a clue of the improvement gained by restraining the set of features. We will plot the curves obtained for silhouette selection, then for both silhouette and label score.

```python
df_x = remove_feat_silhouette(df_x_load)
silhouette_curve = []


for i in range (0, 12):
    df_x = remove_feat_silhouette(df_x)
    kmeans_model = KMeans(n_clusters=2, random_state=0).fit(df_x)
    silhouette_curve.append(silhouette(df_x, kmeans_model.labels_))


df_x.drop('age', inplace=True,axis=1,errors='ignore')
kmeans_model = KMeans(n_clusters=2, random_state=0).fit(df_x)
silhouette_curve.append(silhouette(df_x, kmeans_model.labels_))


print_scores(df_x, df_y, kmeans_model.labels_)

axes = plt.gca()
axes.set_ylim([0.45,0.75])
plt.plot(silhouette_curve)
```

```python
df_x = remove_feat_score(df_x_load)
score_curve = []

for i in range (0, 12):
    df_x = remove_feat_score(df_x)
    kmeans_model = KMeans(n_clusters=2, random_state=0).fit(df_x)
    score_curve.append(score(df_y, kmeans_model.labels_))

df_x.drop('allCards', inplace=True,axis=1,errors='ignore')
kmeans_model = KMeans(n_clusters=2, random_state=0).fit(df_x)
score_curve.append(score(df_y, kmeans_model.labels_))

print_scores(df_x, df_y, kmeans_model.labels_)

axes = plt.gca()
axes.set_ylim([0.45,0.75])
plt.plot(silhouette_curve)
plt.plot(score_curve)
```

### Results
In both case, the curves stay rather flat until three features are left : 'games, 'age' and 'allCards', which yields the best results for both silhouette and score. Although refining with respect to score or silhouette does not give the same "path" of eliminations through the features, the three left are the same in both case. With less than three features, score goes completely off, while silhouette goes down when 2 are left, and back to the same score when only one feature is kept. It is quite logical but meaningless for interpretation with respect to the data.

We will now run one more time our feature removing algorithm to trim df_x down to the best fit.

```python
df_x = remove_feat_silhouette(df_x_load)
silhouette_curve = []


for i in range (0, 11):
    df_x = remove_feat_silhouette(df_x)
    kmeans_model = KMeans(n_clusters=2, random_state=0).fit(df_x)
    silhouette_curve.append(silhouette(df_x, kmeans_model.labels_))



kmeans_model = KMeans(n_clusters=2, random_state=0).fit(df_x)
print_scores(df_x, df_y, kmeans_model.labels_)

df_x.columns
```

## Conclusion

As we see, despite having silhouette score around 0.7, visual inspection shows a rather shapeless distribution of data sectioned in two by the algorithm. We could not properly find a configuration of features with high silhouette where players with dark and light skin colors belong to different clusters. Having silhouette close to 0.7 imply that skin color is completely mixed in both clusters. We can therefore conclude that there might be latent factors leading to clustering (a simple interpretion would suggest that our clusters represent young players who played not so many games and didn't (couldn't) get so many yellow or red cards, against older player who potentially got more cards along more games...).

We can thus conclude that either we missed the right combination of features that could isolate two clusters where players of different skin color would be separated, or, as happened with classification, we could conclude that referee data itself does not allow us such clustering - does not introduce such bias - which would be a pretty happy conclusion to our analysis !
